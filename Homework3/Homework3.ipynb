```python
def digitOddEven(n : int):
    OddCnt = 0
    EvenCnt = 0
    if n == 0:
        EvenCnt += 1
    while n > 0:
        if (n % 10) % 2:
            OddCnt += 1
        else:
            EvenCnt += 1
        n //= 10
    return (OddCnt, EvenCnt)

n = int(input())
print(f'le: {digitOddEven(n)[0]}')
print(f'chan: {digitOddEven(n)[1]}')
```

```python
def check(n : int):
    tmp = 1
    cnt = 1
    while tmp <= n:
        if tmp == n:
            return True
            exit()
        else:
            tmp = 3**cnt
            cnt += 1
    return False

n = int(input())
if check(n):
    print("n la dang 3^k")
else:
    print("n ko la dang 3^k")
```

```python
def checkPrime(n : int):
    if n <= 1:
        return False
    for i in range (int(n ** 0.5) + 1):
        if i <= 1 or i >= n:
            continue
        if n % i == 0:
            return False
    return True

def digitPrimeCnt(n : int):
    cnt = 0
    if n == 0:
        pass
    while n > 0:
        cnt += checkPrime(n%10)
        n //= 10
    return cnt

n = int(input())
print(digitPrimeCnt(n))
```
