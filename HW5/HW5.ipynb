W5A3

```python
def checkPrime(n : int):
    if n <= 1:
        return False
    for i in range (int(n ** 0.5) + 1):
        if i <= 1 or i >= n:
            continue
        if n % i == 0:
            return False
    return True

n = int(input())
print(checkPrime(n))
```

W5A4

```python
def divisorSum(n : int):
    sum = 0
    for i in range (1, int(n ** 0.5) + 1):
        if n % i == 0:
            sum += i + n//i
            if i == n//i:
                sum -= i
    sum -= n
    return sum

n = int(input())
print(divisorSum(n) == n)
```

W5A6

```python
def factorial(n : input):
    fac = 1
    for i in range(1, n+1):
        fac *= i
    return fac

n = int(input())
print(factorial(n))
```

```python
a, b = map(float, input().split())
op = input()

match op:
    case "+":
        print(round(a+b, 2))
    case "-":
        print(round(a-b, 2))
    case "*":
        print(round(a*b, 2))
    case "/":
        print(round(a/b, 2))
```

W5A9

```python
def digitSum(n : int):
    sum = 0
    if n == 0:
        sum += 0
    while n > 0:
        sum += n % 10
        n //= 10
    return sum

n = int(input())
print(digitSum(n))
```
