```python
l1 = list(map(int, input().split()))
l2 = list(map(int, input().split()))
dic1 = {}
dic2 = {}
dic3 = {}
for i in range(0, len(l1), 2):
    dic1[l1[i]] = l1[i+1]

for i in range(0, len(l2), 2):
    dic2[l2[i]] = l2[i+1]

for key in dic1:
    if key in dic2:
        dic3[key] = dic1[key] + dic2[key]
        dic2.pop(key)
    else:
        dic3[key] = dic1[key]

for key in dic2:
    if key in dic1:
        dic3[key] = dic1[key] + dic2[key]
        dic1.pop(key)
    else:
        dic3[key] = dic2[key]

print(dic3)
```

W6A10

```python
li = list(map(int, input().split()))
k = int(input())
freq = {}

lip = set()

for i in li:
    if (k-i) in freq:
        lip.add((min(i, k-i), max(i, k-i)))
    freq[i] = freq.get(i, 0) + 1

print(lip)
```

W6A20

```python
def kpos(li : list, k : int):
    for pos in range (len(li)):
        if li[pos] == k:
            return pos+1
    return -1

li = list(map(int, input().split()))
k = int(input())
print(kpos(li, k))
```
